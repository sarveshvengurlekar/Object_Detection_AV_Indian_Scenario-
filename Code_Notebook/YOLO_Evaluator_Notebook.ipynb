# **Step 2 : Model_Evalution** 

```python
from ultralytics import YOLO

# Load model (explicit task)
model = YOLO(
    "/kaggle/input/models/sarveshvengurlekar1/yolo-sota-model-v1/pytorch/default/1/20best.pt",
    task="detect"
)

# Run validation
metrics = model.val(
    data="/kaggle/working/mydata-9/data.yaml",
    imgsz=640,
    batch=16,
    device=0,        # single GPU only
    conf=0.001,
    iou=0.7,
    max_det=1000,
    half=True,
    split="val",
    save_json=True,
    plots=True,
    verbose=True,
    visualize=True,     # optional (slow)
    augment=True,
)

# Print metrics
print("mAP50-95:", metrics.box.map)
print("mAP50:", metrics.box.map50)
print("mAP75:", metrics.box.map75)
print("Per-class mAP:", metrics.box.maps)

# Confusion matrix as DataFrame
df_cm = metrics.confusion_matrix.to_df()
print(df_cm)
```

# **Step 3 : CSV File Creation**

```python
import os

# Convert metrics to CSV string
val_csv = metrics.to_csv()
print(val_csv)

# Define output path
dir_path = "/kaggle/working/runs/detect"
csv_filename = "5_SOTA_V1_val_results.csv"

# Ensure directory exists
os.makedirs(dir_path, exist_ok=True)

# Full file path
full_path = os.path.join(dir_path, csv_filename)

# Write CSV
with open(full_path, "w") as f:
    f.write(val_csv)

print("Saved at:", full_path)
```
